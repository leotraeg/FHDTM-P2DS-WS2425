#Data Preprocessing

##Data Cleaning

```python
#Pandas read function casts club_avg_age as object -> cast data type for club_avg_age as float
df_players.club_avg_age = df_players.club_avg_age.str.replace(",",".").astype(float)
```

```python
#club_stadium_seats non numeric format (e.g. 55.017 Plätze) -> transform club_stadium_seats as int (e.g. 55017)
df_players.club_stadium_seats = df_players.club_stadium_seats.apply(lambda x: int(float(x.split()[0])*1000))
```

```python
#club_current_transfer_balance non numeric format (e.g. -79,10 Mio. €) -> transform club_current_transfer_balance as float (e.g. -79100000.00)
def get_club_transfer_balance_numeric(club_transfer_balance):
    if(type(club_transfer_balance) == float):
      return club_transfer_balance
    else:
      club_transfer_balance_arr = club_transfer_balance.split(" ")
      #"-128,00 Mio. €".split(" ") --> ['-128,00', 'Mio.', '€']
      value = club_transfer_balance_arr[0].replace(",",".").replace("+","")
      return float(float(value)*1000000)
```

```python
df_players.club_current_transfer_balance = df_players.club_current_transfer_balance.apply(get_club_transfer_balance_numeric)
```

```python
#player_number non numeric format as some players have no number (e.g. "-") -> set players without number = 0
df_players.player_number = df_players.player_number.replace("-", 0).astype(int)
```

```python
#player_dob_date contains dob and age
def get_player_age(player_dob):
    player_dob_arr = player_dob.split(" ")
    #"06.12.1997 (25)".split(" ") --> ['06.12.1997', '(25)']
    age = player_dob_arr[1].replace("(","").replace(")","")
    return age
```

```python
df_players["player_age"] = df_players.player_dob.apply(get_player_age).astype(int)
```

```python
#player_dob_date contains dob and age
def get_player_dob_date(player_dob):
    if(type(player_dob) == datetime.date):
      return player_dob
    else:
      player_dob_arr = player_dob.split(" ")
      #"06.12.1997 (25)".split(" ") --> ['06.12.1997', '(25)']
      dob_arr = player_dob_arr[0].split(".")
      return date(int(dob_arr[2]), int(dob_arr[1]), int(dob_arr[0]))
```

```python
df_players.player_dob = df_players.player_dob.apply(get_player_dob_date)
```

```python
#player_value non numeric format (e.g. -79,10 Mio. €) -> transform player_value as float (e.g. -79100000.00)
def get_player_value_numeric(player_value):
    if(type(player_value) == float):
      return player_value
    else:
      player_value_arr = player_value.split(" ")
      #"150 Tsd. €".split(" ") --> ['150,00', 'Tsd.', '€']
      value = player_value_arr[0].replace(",",".")
      unit = 1000 if "Tsd." in player_value_arr[1] else 1000000
      return float(float(value)*unit)
```

```python
df_players.player_value = df_players.player_value.apply(get_player_value_numeric)
```

##Data Transformation

```python
#attribute construction of international ratio of clubs
def club_internationality(number_player, number_foreign_players):
  foreign_ratio = number_foreign_players / number_player
  if foreign_ratio > 0.85:
    return "extremly international"
  elif foreign_ratio > 0.2:
    return "very international"
  else:
    return "international"
```

```python
df_players["club_internationality"] = (df_players.apply(lambda x: club_internationality(x.club_number_player, x.club_number_foreign_players), axis=1))
```

```python
#numeric scaling of player positions
df_players["player_position_scale"] = df_players.player_position.map({'Torwart': 1, 'Abwehr': 2, 'Mittelfeld': 3, 'Sturm':4, np.NaN:0})
```

```python
#club leauge is linked to a specific country -> construct attribute of club country
df_players.club_league.unique()
```

```python
dict_league_country = {'Premier League':"England",
                       'Ligue 1':"Frankreich",
                       'LaLiga':"Spanien",
                       'Bundesliga':"Deutschland",
                       'Serie A':"Italien",
                       'Championship':"England",
                       'Liga Portugal':"Portugal",
                       'Süper Lig':"Türkei",
                       'Eredivisie':"Niederlande",
                       'Saudi Pro League':"Saudi-Arabien",
                       'Campeonato Brasileiro Série A':"Brasilien",
                       'Premier Liga':"Russland",
                       'Jupiler Pro League':"Belgien",
                       'Scottish Premiership':"Schottland",
                       'SuperSport HNL':"Kroatien",
                       'Liga Profesional de Fútbol':"Argentinien"}
```

```python
df_players["club_country"] = df_players.club_league.map(dict_league_country)
```

```python
df_players["player_international"] = (df_players.apply(lambda x: 1 if x.player_country != x.club_country else 0, axis=1))
```
